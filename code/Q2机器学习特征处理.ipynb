```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import math
```

```python
#1.1 读取数据
data = pd.read_excel('../data/特征提取ffffffinal.xlsx')

#提取需要处理的特征，其余特征扔掉(包括单词本身)
df2 = data[['1 try','2 tries','3 tries','4 tries','5 tries','6 tries','7 or more tries (X)',
'tags','week','char1','char2','char3','char4','char5','Word feq','Vowels nums','Combined nums','Letter feq','Letter rep']]
```

```python
# 1.2 特征工程

#添加时间特征
df2['month'] = data['Date'].dt.month
df2['day'] = data['Date'].dt.day

#添加定序编码
from sklearn.preprocessing import LabelEncoder
#训练
chars = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h',
       'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v',
       'w', 'x', 'y', 'z']
label1 = LabelEncoder()
label1.fit(chars)

label2 = LabelEncoder()
label2.fit(['NN','JJ','others'])
#应用
df2['char1'] = label1.transform(df2['char1'])
df2['char2'] = label1.transform(df2['char2'])
df2['char3'] = label1.transform(df2['char3'])
df2['char4'] = label1.transform(df2['char4'])
df2['char5'] = label1.transform(df2['char5'])
df2['tags'] = label2.transform(df2['tags'])
df2
```

```python
df2.to_excel('../data/机器学习特征工程_V3.xlsx',index = False)
```
