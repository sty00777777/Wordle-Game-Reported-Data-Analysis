```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False    
import seaborn as sns
```

```python
#读取数据
data = pd.read_excel('../data/清洗数据.xlsx')
data.head()
```

```python
data1 =data[['Number of  reported results','Number in hard mode','1 try','2 tries','3 tries','4 tries','5 tries',
'6 tries','7 or more tries (X)']]
```

```python
# Numbers 的相关性图
plt.figure(figsize = (10,10))
sns.pairplot(data1[['Number of  reported results','Number in hard mode']])
#plt.savefig('../latex/pic/Numbers相关性图.png',dpi = 200)
plt.show()
```

```python
# try 的相关性图
plt.figure(figsize = (10,10))
sns.pairplot(data1[['1 try','2 tries','3 tries','4 tries','5 tries','6 tries','7 or more tries (X)']],kind='reg')
#plt.savefig('../latex/pic/tries相关性图reg.png',dpi = 200)
plt.show()
```
